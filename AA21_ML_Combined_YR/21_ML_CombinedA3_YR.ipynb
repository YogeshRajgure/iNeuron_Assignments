# Assignment3

In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster,datasets
from scipy import misc
```

```python
raw_img = misc.face(gray = True)
plt.figure(figsize=(10,5))
plt.imshow(raw_img, plt.cm.gray)
```

```python
# getting rows and columns of image
```

```python
rows, columns = raw_img.shape[0], raw_img.shape[1]
```

```python
image = raw_img.reshape(rows*columns, 1) 
```

```python
# compressing Image using kmeans
```

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters = 5)
kmean.fit(image)
```

```python
# getting clusters and labels for image
```

```python
cluster = np.asarray(kmean.cluster_centers_)
cluster
```

```python
labels = np.asarray(kmean.labels_)
labels
```

```python
labels = labels.reshape(rows, columns)
```
