In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.\
The raw image is available in spicy.misc package with the name face.

```python
import pandas as pd
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Visualising image before compressing
```

```python
racoon = misc.face(gray = True)
plt.figure(figsize=(10,5))
plt.imshow(racoon, plt.cm.gray)
```

```python
# grtting rows and columns of image
```

```python
rows = racoon.shape[0]
columns = racoon.shape[1]
image = racoon.reshape(rows*columns, 1)
```

# Compressing Image

```python
# using k means to fit the image
```

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters = 5)
kmean.fit(image)
```

```python
# getting cluster and labels for the image
```

```python
clusters = np.asarray(kmean.cluster_centers_)
clusters
```

```python
labels = np.asarray(kmean.labels_)
labels
```

```python
labels = labels.reshape(rows, columns)
labels
```

```python
# save the compressed image
```

```python
plt.imsave('/content/drive/MyDrive/Colab Notebooks/compressed_racoon.png', labels)
```
